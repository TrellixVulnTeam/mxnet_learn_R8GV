Example 6

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

for ax in axes:
    ax.plot(x, y, 'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('example 6')
#防止图片堆叠
fig.tight_layout()
```

Example 8

```python
fig, ax = plt.subplots()

ax.plot(x, x ** 2, label = 'y = x ** 2')
ax.plot(x, x ** 3, label = 'y = x ** 3')
ax.legend(loc=2) # 将label上的提示放置在图片的哪一块位置，2表示左上角
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('example 8')
```

```python
# 支持laTex
fig, ax = plt.subplots()

ax.plot(x, x ** 2, label = r'$y = \alpha^2$')
ax.plot(x, x ** 3, label = r'$y = \alpha^3$')
ax.legend(loc=2) # 将label上的提示放置在图片的哪一块位置，2表示左上角
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('example 8')
```

Example 10

```python
# 自定义标签
fig, ax = plt.subplots(figsize = (10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize = 18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(['$%.1f$' % y for y in yticks], fontsize = 18)
```

Example 11

```python
#绘制网格
fig, ax = plt.subplots(1, 2, figsize = (10, 3))

ax[0].plot(x, x**2, x, x**3, lw=5)
ax[0].grid(True)

ax[1].plot(x, x**2, x, x**3, lw=2)
ax[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

Example 12

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data', 0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))

xx = np.linspace(-0.75, 1.,100)
ax.plot(xx, xx**3)
```

Example 13

```python
# 绘制其他类型的图表
n = np.array([0, 1, 2, 3, 4, 5])

fig, ax = plt.subplots(1, 4, figsize=(14, 4))
#散点图
ax[0].scatter(xx, xx + 0.25 * np.random.randn(len(xx)))
ax[0].set_title('scatter')

ax[1].step(n, n**2, lw=2)
ax[1].set_title('step')

ax[2].bar(n, n**2, align='center', width=0.5, alpha=0.5)
ax[2].set_title('bar')

ax[3].fill_between(x, x**2, x**3, color='green', alpha=0.5)
ax[3].set_title('fill_between')
```

```python
#极坐标
fig = plt.figure()
ax = fig.add_axes([0.0, 0.0, .6, .6], polar=True)
t = np.linspace(0, 2 * np.pi, 100)
ax.plot(t, t, color='blue', lw = 3)
```

```python
#直方图
n = np.random.rand(100000)
fig, ax = plt.subplots(1, 2, figsize=(14, 4))

ax[0].hist(n)
ax[0].set_title('default histogram')
ax[0].set_xlim((min(n), max(n)))

ax[1].hist(n, cumulative=True, bins=50)
ax[1].set_title('cumulative detailed histogram')
ax[1].set_xlim(min(n), max(n))
```
